### Resampling the Data

```python
# Resampling the data in order to average the rows for every hour in the dataset
# For every hour there is 180 values corresponding to every 20s and we are taking the average of these values
df['NewDateTime'] = pd.to_datetime(df['date'])
df.index = df['NewDateTime']
df = df.resample('H').mean()
df.head()
```

```python
df.info()
```

```python
# Replacing the empty rows with NaN values in order to drop them later
df = df.replace('', np.nan)
df.loc[df.isna().any(axis=1)]
```

```python
# Removing the missing values
df.dropna(inplace = True)
```

### Removal of Outliers

```python
# Using the IQR score to retrive and filter out the outliers by keeping only valid values
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
df_out = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df = df_out
df.shape
```

```python
# Removing % Iron Concentrate attribute due to its high correlation with the target variable % Silica Concentrate
df.drop('% Iron Concentrate', axis = 1, inplace = True)
```

### Removing Highly Correlated Features

```python
import numpy as np

# Create correlation matrix
corr_matrix = df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k = 1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.75
to_drop = [column for column in upper.columns if any(upper[column] > 0.75)]

# Drop features 
df.drop(df[to_drop], axis = 1, inplace = True)

df.head()
```

### Feature Engineering

```python
# Taking the average of the remaining Flotation Air Flows and the FLotation Levels and created two new columns 
df['Ave_Flot_Air_Flow'] = df[['Flotation Column 01 Air Flow', 'Flotation Column 02 Air Flow', 
                              'Flotation Column 04 Air Flow', 'Flotation Column 05 Air Flow', 
                              'Flotation Column 06 Air Flow', 'Flotation Column 07 Air Flow']].mean(axis = 1)

df['Ave_Flot_Level'] = df[['Flotation Column 01 Level', 'Flotation Column 03 Level', 
                           'Flotation Column 04 Level']].mean(axis = 1)

df.drop(['Flotation Column 01 Air Flow', 'Flotation Column 02 Air Flow', 'Flotation Column 04 Air Flow', 
         'Flotation Column 05 Air Flow', 'Flotation Column 06 Air Flow', 'Flotation Column 07 Air Flow',
         'Flotation Column 01 Level', 'Flotation Column 03 Level', 'Flotation Column 04 Level'], axis = 1, inplace = True)

df.head()
```

```python
# Reordering the columns in the dataset
cols = df.columns.tolist()
order = cols[-2:] + cols[:-2]
df = df[order]
```
